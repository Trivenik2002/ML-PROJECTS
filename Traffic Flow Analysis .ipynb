1.IMPORT PYTHON LIBRARIES

```python
import numpy as np
import pandas as pd
```

2.LOAD THE DATASET

```python
data=pd.read_csv("traffic.csv")
data
```

4.DATA IMPUTATION

```python
data["Day of the week"]=data["Day of the week"].fillna(data["Day of the week"].mode()[0])
data["CarCount"]=data["CarCount"].fillna(data["CarCount"].median())
data["Traffic Situation"]=data["Traffic Situation"].fillna(data["Traffic Situation"].mode()[0])
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
cat_features=[feature for feature in data.columns if data[feature].dtype=='object']
cat_features
```

5.DATA ENCODING

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for cat_features in data.columns:
   data[cat_features]=le.fit_transform(data[cat_features])
print(data.head())
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x
```

```python
y
```

6.DATA NORMALIZATION

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit_transform(x)
```

7.SPLIT THE DATA

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
y_test.head
```

8.ALGORITHMS

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
y_pred
```

9.MODEL PERFOMENCE

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
svm=SVC()
svm.fit(x_train,y_train)
```

```python
y_pred1=svm.predict(x_test)
y_pred1
```
