```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df=pd.read_csv("synthetic_coffee_health_10000.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Health_Issues'].value_counts()
```

```python
df['Health_Issues']=df['Health_Issues'].fillna(df['Health_Issues'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
cat_features=[feature for feature in df.columns if df[feature].dtype=='object']
cat_features
```

```python
le=LabelEncoder()
for cat_features in df.columns:
    df[cat_features]=le.fit_transform(df[cat_features])
print(df.head())    
```

```python
x=df.drop(columns='Health_Issues',axis=1)
```

```python
y=df['Health_Issues']
```

```python
x
```

```python
y
```

```python
s=StandardScaler()
s.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
y_pred=dtc.predict(x_test)
y_pred
```

```python
print(accuracy_score(y_test,y_pred)*100)
```

```python
print(confusion_matrix(y_test,y_pred))
```
